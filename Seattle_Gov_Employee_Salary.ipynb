```python
import pandas as pd
from sodapy import Socrata
import numpy as np
```

```python
employees = Socrata("data.seattle.gov", None).get("s7qb-vteq")
department = [employee['department'] for employee in employees]
first_name = [employee['first_name'] for employee in employees]
last_name = [employee['last_name'] for employee in employees]
job_title = [employee['job_title'] for employee in employees]
hourly_rate = [employee['hourly_rate'] for employee in employees]

wage_columns = ['department','first_name','last_name','job_title','hourly_rate']
wage_data = [department,first_name,last_name,job_title,hourly_rate]
wage = pd.DataFrame(dict(list(zip(wage_columns, wage_data))))
wage['full_name'] = wage['first_name'] + ' ' + wage['last_name']
wage.head()
```

```python
wage.describe()
```

```python
wage.info()
```

```python
wage['hourly_rate'] = wage['hourly_rate'].astype(float)
wage.info()
```

```python
wage.groupby('department').agg(max_rate = ('hourly_rate','max'),
                              min_rate = ('hourly_rate','min'),
                              avg_rate = ('hourly_rate', 'mean'),
                              total_employees = ('hourly_rate','count'),
                              unique_jobtitle = ('job_title','nunique')).sort_values(['avg_rate','department'],ascending = [False, True])                          
```

```python
wage.sort_values(by= ['department','hourly_rate'], ascending = [True, False], inplace = True)
wage = wage.reset_index(drop = True)
wage['wage_rank'] = wage.groupby(['department'])['hourly_rate'].rank(method='dense', ascending = False) 
wage['wage_rank'] = wage['wage_rank'].astype(int)
wage
```

```python
# employees with the highest hourly_rate for each empployee
wage[wage['wage_rank'] == 1].reset_index(drop = True)
```

```python
wage['max_hourly_rate'] = wage['hourly_rate'].max()
#wage['max_hourly_rate_by_department'] = wage.groupby(['department'])['hourly_rate'].transform('max').round(2)
wage['max_hourly_rate_by_department'] = wage.groupby(['department'])['hourly_rate'].transform('max')
wage['average_hourly_rate'] = wage['hourly_rate'].mean()
wage['average_hourly_rate_by_department'] = wage.groupby(['department'])['hourly_rate'].transform('mean')
wage
```
